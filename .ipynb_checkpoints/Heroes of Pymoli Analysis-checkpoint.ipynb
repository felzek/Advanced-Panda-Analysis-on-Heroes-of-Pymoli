```python
import pandas as pd
import numpy as np 
import os



datasource = os.path.join("Source_data","purchase_data.json")
datasource_2 = os.path.join("Source_data","purchase_data2.json")

purchase_data_df= pd.read_json(datasource)
purchase_data_df_2 = pd.read_json(datasource_2)


merged_data_dfa = pd.merge(purchase_data_df,purchase_data_df_2, on = ["Age","Gender","Item ID","Item Name","Price","SN"], how ="outer" )
merged_data_df = merged_data_dfa.drop_duplicates()

```

```python
total_players = merged_data_df["SN"].nunique()
total_players_df = pd.DataFrame({"Total Players":[total_players]})
total_players_df.head()

```

<h1 style="font-size:100;font-weight:600;">Purchasing Analysis</h1>

z## Player Count
purchase_data_df.columns

```python
num_unique_items = merged_data_df["Item ID"].nunique()
average_purchase_price = merged_data_df["Price"].mean()
total_number_of_purchases= merged_data_df["Item ID"].count()
total_revenue = merged_data_df["Price"].sum()

purchasing_analysis = pd.DataFrame({"Number of Unique Items":[num_unique_items],
                                    "Average Price":"$" + str(round(average_purchase_price,2)),
                                    "Number of Purchases":[total_number_of_purchases],
                                    "Total Revenue":"$" + str(round(total_revenue,2)) })
purchasing_analysis

```

<h1 style="font-size:100;font-weight:600;">Gender Demographics</h1>

```python
total_gender = merged_data_df["Gender"].count()
male = merged_data_df["Gender"].value_counts()["Male"]
female = merged_data_df["Gender"].value_counts()["Female"]
non_gender = total_gender - male - female

percentage_male_count = male / total_gender * 100
percentage_female_count = female / total_gender * 100
percentage_nonspecific_count = non_gender / total_gender * 100

gender_demographics_df = pd.DataFrame({"Male": [male,percentage_male_count] ,
                                       "Female":[female,percentage_female_count],
                                       "Non-Specific":[non_gender,percentage_nonspecific_count]
                                       
                                      
                                      },index = ["Count","Percentage of Players"])
gender_demographics_df.transpose()
```

<h1 style="font-size:100;font-weight:600;">Purchasing Analysis (Gender)</h1>

```python
male_purchase = merged_data_df.loc[merged_data_df["Gender"] == "Male", :]
female_purchase = merged_data_df.loc[merged_data_df["Gender"] == "Female",:]
non_specific_purchase = merged_data_df.loc[merged_data_df["Gender"] == "Other / Non-Disclosed",:]

#Purchase Count

male_purchase_count = male_purchase["Item ID"].count()
female_purchase_count = female_purchase["Item ID"].count()
non_specific_purchase_count = non_specific_purchase["Item ID"].count()

#Average Purchase Price

male_average_price = round(male_purchase["Price"].mean(),2)
female_purchase_price = round(female_purchase["Price"].mean(),2)
non_specific_purchase_price = round(non_specific_purchase["Price"].mean(),2)

#Total Purchase Value
male_total_purchase = round(male_purchase["Price"].sum(),2)
female_total_purchase = round(female_purchase["Price"].sum(),2)
non_specific_total_purchase = round(non_specific_purchase["Price"].sum(),2)

#Normalized Totals
male_normalized = male_total_purchase/male
female_normalized = female_total_purchase/female
non_specific_normalized = non_specific_total_purchase/non_gender

purchasing_analysis_df = pd.DataFrame({"Purchase Count":[male_purchase_count,female_purchase_count,non_specific_purchase_count],
                                       "Average Purchase Price":["$" + str(male_average_price),"$" + str(female_purchase_price),"$" + str(non_specific_purchase_price)],
                                       "Total Purchase value":["$" + str(male_total_purchase),"$"+ str(female_total_purchase),"$" + str(non_specific_total_purchase)],
                                       "Normalized Totals":["$" + str(male_normalized),"$" + str(female_normalized),non_specific_normalized]
                                    },index=["Male","Famale","Other/ Non-Disclosed"])

purchasing_analysis_df
```

<h1 style="font-size:100;font-weight:600;">Age Demographics</h1>

```python
age_group_names = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40-44","45-49"]
bins = [0,9,14,19,24,29,34,39,44,49]

merged_data_df_2 = pd.DataFrame(merged_data_df)
merged_data_df_2["Age Group"] = pd.cut(merged_data_df["Age"],bins,labels=age_group_names)
merged_data_df_2 = merged_data_df_2.groupby("Age Group", as_index=True)



age_count = merged_data_df_2["Age Group"].count()
age_total = merged_data_df["Age"].count()
age_percentage = age_count/age_total*100

age_demographics_df = pd.DataFrame({"Age Count":age_count,"Age Percentage": age_percentage})

age_demographics_df.head(10)

```

```python
#Age Analysis


age_purchase_count = merged_data_df_2["Item ID"].count()
age_average_purchase_price = merged_data_df_2["Price"].mean() 
age_total_purchase_value = merged_data_df_2["Price"].sum()
age_normalized_total = age_total_purchase_value/age_purchase_count

age_demographics_analysis = pd.DataFrame({"Purchase Count":age_purchase_count,
                             "Average Purchase Price":age_average_purchase_price,
                              "Total Purchase Value":age_total_purchase_value,
                              "Normalized Totals":age_normalized_total
                            })

age_demographics_analysis.head(10)
```

```python

top_SN = merged_data_df.groupby("SN").Price.agg(("count","sum","mean"))
top_SN.sort_values('sum',ascending=False,inplace = True)
top_SN.rename(columns={"count":"Purhcase Count","sum":"Total Purchase Value","mean":"Average Purchase Price"},inplace=True)
top_SN["Average Purchase Price"]="$" + (round(top_SN["Average Purchase Price"],2)).astype(str)
top_SN["Total Purchase Value"]="$" + (round(top_SN["Total Purchase Value"],2)).astype(str)
top_SN.iloc[0:5]

```
